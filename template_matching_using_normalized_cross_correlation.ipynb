# Importing Required Packages

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

# Inserting and Displaing Image and Template

```python
test_image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Untitled folder/test_5.jpg")
template = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Untitled folder/template_5.jpg')
(tH, tW) = template.shape[:2]

# displaying the  image and template in screen
print("Test Image")
cv2_imshow(test_image)
print("template image")
cv2_imshow(template)
```

# Converting Both Image and Template to Grayscale and Performing Template Matching

```python
# converting both the image and template to grayscale
imageGray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
templateGray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
# performing template matching
print("[INFO]performing template matching...")
result = cv2.matchTemplate(imageGray, templateGray,	cv2.TM_CCORR_NORMED)
```

```python
imageGray.shape, templateGray.shape
```

```python
threshold = 0.99

(yCoords, xCoords) = np.where(threshold < result)
clone = test_image.copy()
print("matched locationS")

for (x, y) in zip(xCoords, yCoords):
	cv2.rectangle(clone, (x, y), (x + tW, y + tH), (255, 0, 0))
cv2_imshow(clone)
```
